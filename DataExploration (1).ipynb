```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define labels for the embarked values
embarked_labels = {1: 'Cherbourg', 2: 'Queenstown', 3: 'Southampton'}

# Group by 'Embarked' and count the number of survivors
survivors_by_embarked = train_df.groupby(['Embarked'], as_index=False)['Survived'].count()

# Rename the 'Embarked' values with labels
survivors_by_embarked['Embarked'] = survivors_by_embarked['Embarked'].map(embarked_labels)

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Embarked', y='Survived', data=survivors_by_embarked)
plt.title('Count of Survivors by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Count of Survivors')
plt.show()
```
